```python
unique_features = []
def extract_unique_features(features):
    global unique_features
    features = list(map(str.strip,features.split(",")))
    for feature in features:
        if feature not in unique_features:
            unique_features.append(feature)
        return features
        
```

```python
def update_feature_availability(data_row):
    options = data_row.Options
    options = list(map(str.strip,data_row.Options.split(",")))
    for feature in unique_features:
        if feature in options:
            data_row[feature] = "Available"
    return data_row
```
